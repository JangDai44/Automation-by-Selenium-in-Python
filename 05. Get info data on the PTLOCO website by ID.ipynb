```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
options = webdriver.FirefoxOptions()
options.headless = True
driver = webdriver.Firefox(options=options)

# Website used
driver.get("https://photoloco.com.au/")
```

```python
dic= {
    'ID': "None",
    'Name': "None",
    'Location': "None",
    'State' : "None"
    }
list_dic=[]
info=[]
list_id=[]

# Here select the codes 102x and 103x. If you want to add a list, change the number and for loop to suit your needs
for j in range(2,4):
    driver.get("https://photoloco.com.au/?s="+"PL10"+str(j)+"&post_type=product&type_aws=true&id=1&filter=1")
    for k in driver.find_elements(By.XPATH, "/html/body/main/div[1]/div/section[1]/div/div[1]"):
        for i in k.find_elements(By.XPATH, "./div"):
            name= i.find_element(By.XPATH, ".//div[2]/div[1]/a[1]").text
            dic["ID"]= name.split()[0]
            list_id.append(dic["ID"])
            dic["Name"]= name
            count=0
            for l in i.find_elements(By.XPATH, ".//div[2]/div[2]/a"):
                count +=1
            if count==2:
                dic["Location"]= i.find_element(By.XPATH, ".//div[2]/div[2]/a[1]").text
                dic["State"]= i.find_element(By.XPATH, ".//div[2]/div[2]/a[2]").text
            else:
                dic["State"]= i.find_element(By.XPATH, ".//div[2]/div[2]/a").text
            # Add a Dictionary to a List
            list_dic.append(dic)
            dic= {
                    'ID': "None",
                    'Name': "None",
                    'Location': "None",
                    'State' : "None"
                    }
print(list_dic)
print(list_id)
```

```python
# convert list into csv file for display
import pandas as pd

df = pd.DataFrame(list_dic,columns=["ID", "Name", "Location", "State"])
df= df.sort_values(by=["ID"]) # Sort values by ID
df.to_csv('list.csv', index=False)
file= pd.read_csv('list.csv')
file
```
