# 02. Get List File Name in Zipfile/Rar File

```python
! pip install rarfile

import rarfile
import zipfile
```

```python
# Declare Dictionary and List
dic= {
    'File Zip': "None",
    'File Name': "None"
    }
info=[]

# find Rar file in Folder link
url= "/content/drive/MyDrive/Colab Notebooks/2023"
for i in os.listdir(url):
  if i[-4:]== ".rar":
    stories_rar = rarfile.RarFile(url+"/"+i)
    for file_ in stories_rar.namelist():
      if file_[-4:] == ".jpg" or file_[-4:] == ".png" or file_[-5:] == ".jpeg":
        dic['File Zip']=file_.split("/")[0]
        dic['File Name']= file_.split("/")[-1]
        info.append(dic)
        dic= {
    'File Zip': "None",
    'File Name': "None"
    }

# find Zip file in Folder link
  if i[-4:]== ".zip":
    stories_zip = zipfile.ZipFile(url+"/"+i)
    for file_ in stories_zip.namelist():
      if file_[-4:] == ".jpg" or file_[-4:] == ".png" or file_[-5:] == ".jpeg":
        dic['File Zip']=file_.split("/")[0]
        dic['File Name']= file_.split("/")[-1]
        info.append(dic)
        dic= {
    'File Zip': "None",
    'File Name': "None"
    }
```

```python
# Save list to csv file
import pandas as pd

df = pd.DataFrame(info)
df.to_csv('/content/drive/MyDrive/Colab Notebooks/File/file.csv', index=False)
file= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/File/file.csv')
file
```
