```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
# Open Web 
options = webdriver.FirefoxOptions()
options.headless = True
driver = webdriver.Firefox(options=options)
driver.get("<Link administrator WordPress website>")

# Sign in wordpress administrator
user= driver.find_element(By.ID, "user_login").send_keys("<username>")
password= driver.find_element(By.ID, "user_pass").send_keys("<password>")
login= driver.find_element(By.ID, "wp-submit").click()
```

```python
list_info=[]
dic= {
    'OrderID': "None",
    'SKU': "None",
}
for i in range (290441,290489): # check orderID from 290441 to 290489
    driver.get("<Link administrator WordPress website>/post.php?post="+str(i)+"&action=edit")
    dic['OrderID']='#LDF'+str(i)
    list_sku=[]
    for j in driver.find_elements(By.XPATH, '//*[@id="order_line_items"]/tr'):
        sku= j.find_element(By.CLASS_NAME, 'wc-order-item-sku').text
        list_sku.append(sku.split()[-1]) 
    if len(list_sku)>0:
        sku_str=', '.join(map(str,list_sku)) # concatenates the elements of the list into a string
    else:
        sku_str=' '.join(map(str,list_sku))
    dic['SKU']=sku_str
    list_info.append(dic)
    dic= {
    'OrderID': "None",
    'SKU': "None",
}
driver.close()
list_info
```

```python
# convert list into csv file for display to table
import pandas as pd

df = pd.DataFrame(list_info,columns=["OrderID", "SKU"])
df= df.sort_values(by=["OrderID"]) # Sort values by ID
df.to_csv('list_info.csv', index=False)
file= pd.read_csv('list_info.csv')
file
```
